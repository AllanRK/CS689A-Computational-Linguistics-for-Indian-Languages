```python
# import the required libraries ...
import conllu
import operator
with open('hi_hdtb-ud-dev.conllu', mode="r", encoding="utf-8") as data:
    
    # Read the file contents and assign under 'annotations'
    annotations = data.read()

# Use the parse() function to parse the annotations; store under 'sentences'
sentences = conllu.parse(annotations)
```

# Question 1-a

```python
list_of_pos = [] # list of pos tags
list_of_form = [] # list of words
word_pos_tag = [] # list of word---pos tags
form_pos = {}
for i in range(len(sentences)):
    for j in range(len(sentences[i])):
        x = sentences[i][j]['form']
        list_of_form.append(x)
        y = sentences[i][j]['upos']
        list_of_pos.append(y)
        form_pos[x] =y 
        word_pos = x+"---"+y
        word_pos_tag.append(word_pos)
        
```

```python
print(form_pos)
```

```python
# Frequency of POS tags of words
from collections import Counter
count_of_pos_tags = Counter(list_of_pos)
pos_tag_frequency = ( sorted(count_of_pos_tags.items(), key=operator.itemgetter(1),reverse=True))
print(pos_tag_frequency)
```

```python
# Frequency of POS tags of each word
count_word_pos_tag = Counter(word_pos_tag)
word_pos_tag_frequency = ( sorted(count_word_pos_tag.items(), key=operator.itemgetter(1),reverse=True))
print(word_pos_tag_frequency)
```

# Question 1-b

```python
# 50 most frequent words corresponding to each POS tag
set_of_pos_tags = set(list_of_pos)
for pos in set_of_pos_tags:
    list_of_words = []
    for i in range(len(sentences)):
        for j in range(len(sentences[i])):
            word = sentences[i][j]['form']
            word_pos = sentences[i][j]['upos']
            if  word_pos == pos:
                list_of_words.append(word)
    print("------",pos,"-------")
    word_frequency = {}

    for item in list_of_words:
        if item in word_frequency:
            word_frequency[item] += 1
        else:
            word_frequency[item] = 1

        
    import operator
    sorted_word_frequency = ( sorted(word_frequency.items(), key=operator.itemgetter(1),reverse=True))
    print(sorted_word_frequency[:50])
```

# Question 1- e

```python
# function to obtain the head words along with corresponding pos tags
def get_frequency_head_pos(sentences):
    
    list_of_head_pos = []
    list_of_head_word_pos = []
    #word_head_pos = []
    for i in range(len(sentences)):
        for j in range(len(sentences[i])):
            x = sentences[i][j]['misc']['ChunkType']
            if( x == 'head'):
                head_pos = sentences[i][j]['upos']
                list_of_head_pos.append(head_pos)
                list_of_head_word_pos.append(sentences[i][j]['form']+"--"+head_pos)
    return list_of_head_pos,list_of_head_word_pos
```

```python
# frequencies of POS tags corresponding to only head words
list_of_head_pos,list_of_head_word_pos = get_frequency_head_pos(sentences)
count_of_head_pos = Counter(list_of_head_pos)
head_pos_frequency = ( sorted(count_of_head_pos.items(), key=operator.itemgetter(1),reverse=True))
print(head_pos_frequency)
```

```python
# frequencies of POS tags corresponding to each head words
count_of_head_word_pos = Counter(list_of_head_word_pos)
head_word_pos_frequency = ( sorted(count_of_head_word_pos.items(), key=operator.itemgetter(1),reverse=True))
print(head_word_pos_frequency)
```

# Question 1-f

```python
# For each directed POS tag tuples, i.e., <ti, tj>, 
# list the frequencies separately for each relation R as well as total.

list_of_tuples = []
list_of_relation = []
for i in range(len(sentences)):
    for j in range(len(sentences[i])):
        id = sentences[i][j]['head']
        t1 = sentences[i][j]['upos']
        if id != 0:
            t2 = sentences[i][id-1]['upos']
            x = "< "+t1+","+t2+" >"
        else:
            x = t1
        relation = sentences[i][j]['deprel']
        list_of_tuples.append(x)
        list_of_relation.append(relation)  
#for i in range(len(list_of_tuples)):
    #print(list_of_tuples[i],"--------",list_of_relation[i])
```

```python
set_of_tuple_relations = set(list_of_tuples)
dict1 = {}
for j in set_of_tuple_relations:
    
    x = [i for i, e in enumerate(list_of_tuples) if e == j]
    z = []
    for y in x:
        z.append(list_of_relation[y])
    dict1[j] = z
    
#print(dict1)
for j in set_of_tuple_relations:
    m = dict1[j]
    x1 = Counter(m)
    relation_frequency = ( sorted(x1.items(), key=operator.itemgetter(1),reverse=True))
    print(j)
    print("-------Relation Frequency-------")
    print(relation_frequency)
    print("Total Relation Frequency: ",len(m))
    
```
